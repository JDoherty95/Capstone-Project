```python
# import dependencies
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
import hvplot.pandas
from bokeh.models.formatters import BasicTickFormatter
```

```python
# grab btc data from yahoo finance. start at 9/17/2014 since thats when the data starts
btc_df = yf.download("BTC-USD", start="2014-09-17")
btc_df
```

```python
## grab Nasdaq 100 data from yahoo finance. start at 9/17/2014 since thats when the data starts for Bitcoin
ndaq_df = yf.download("NDAQ", start="2014-09-17")
ndaq_df.head()
```

```python
# Plot Bitcoins Historical Price
plt.figure(figsize=(12,5),facecolor='blue')
plt.plot(btc_df.index, btc_df["Close"], label="BTC", color="red")
plt.xlabel("Date",fontweight="bold",size=13)
plt.ylabel("Bitcoin Price",fontweight="bold",size=13)
plt.legend(prop={"size":20})
plt.title("Bitcoin Historical Closing Prices",fontweight="bold",size=20)
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot Bitcoins Historical Price
plt.figure(figsize=(12,5),facecolor='blue')
plt.plot(ndaq_df.index, ndaq_df["Close"], label="NDAQ", color="red")
plt.xlabel("Date",fontweight="bold",size=13)
plt.ylabel("Nasdaq Price",fontweight="bold",size=13)
plt.legend(prop={"size":20})
plt.title("Nasdaq Historical Closing Prices",fontweight="bold",size=20)
plt.xticks(rotation=45)
plt.show()
```

```python
btc_df_2 = btc_df.reset_index()
# Make new columns to find the month&year together, and Month and Year by itself
btc_df_2['Month_Year'] = pd.to_datetime(btc_df_2['Date']).dt.to_period('M')
#btc_df_2['Month'] = pd.DatetimeIndex(btc_df_2['Date']).month
#btc_df_2['Day'] = pd.DatetimeIndex(btc_df_2['Date']).day
btc_df_2 = btc_df_2.drop(columns=["Open","High","Low","Adj Close","Date"],axis=1)
btc_df_2
```

```python
# Over the past four years, show the lowest volume for each month
btc_mean = btc_df_2.groupby("Month_Year").mean()
btc_mean.rename(columns = {"Month_Year":"Avg Close Price"}, inplace = True)
# min_spy_volume = min_spy_volume.drop(columns = ["Date","Year","Date","Month_Year"],axis=1)
# Display data
btc_mean
```

```python
# Plot the bitcoin mean prices
btc_mean_plot = btc_mean.hvplot.bar(x='Month_Year', y='Close', xlabel='Month&Year',frame_width=1000,height=400,ylabel='Bitcoin Close Price',color="red",rot=90,title="Bitcoin's Mean Closing Prices",yformatter=BasicTickFormatter(use_scientific=True))
btc_mean_plot
```

```python
ndaq_df_2 = ndaq_df.reset_index()
# Make new columns to find the month&year together, and Month and Year by itself
ndaq_df_2['Month_Year'] = pd.to_datetime(ndaq_df_2['Date']).dt.to_period('M')
#btc_df_2['Month'] = pd.DatetimeIndex(btc_df_2['Date']).month
#btc_df_2['Day'] = pd.DatetimeIndex(btc_df_2['Date']).day
ndaq_df_2 = ndaq_df_2.drop(columns=["Open","High","Low","Adj Close","Date"],axis=1)
ndaq_df_2
```

```python
# Over the past four years, show the lowest volume for each month
ndaq_mean = ndaq_df_2.groupby("Month_Year").mean()
ndaq_mean.rename(columns = {"Month_Year":"Avg Close Price"}, inplace = True)
# min_spy_volume = min_spy_volume.drop(columns = ["Date","Year","Date","Month_Year"],axis=1)
# Display data
ndaq_mean
```

```python
# Plot the bitcoin mean prices
ndaq_mean_plot = ndaq_mean.hvplot.bar(x='Month_Year', y='Close', xlabel='Month&Year',frame_width=1000,height=400,ylabel='Nasdaq Close Price',color="red",rot=90,title="Nasdaq's Mean Closing Prices",yformatter=BasicTickFormatter(use_scientific=True))
ndaq_mean_plot
```
